```python
selected_column=['country', 'country_long', 'name','capacity_mw','primary_fuel',
                 'other_fuel1','other_fuel2','other_fuel3','commissioning_year','year_of_capacity_data',
                 'generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015',
       'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018',
       'generation_gwh_2019','estimated_generation_gwh'
                 ]
```

```python
len(selected_column)
```

```python
df1=df[selected_column].copy()
```

```python
df1.sample(5)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']= (12,8)
matplotlib.rcParams['figure.facecolor']='#00000000'
```

```python
df1.name.nunique()
```

```python
states_plant=df1.name.value_counts().head(20)
```

```python
states_plant
```

```python
sns.barplot(x=states_plant.index,y=states_plant)
plt.xticks(rotation=90)
plt.title('State Designation')
plt.ylabel('Number of Power Plant')
plt.xlabel('states');
```

```python
main_primary_fuel=df1.primary_fuel.value_counts()*100/df1.primary_fuel.count()
main_primary_fuel
```

```python
sns.barplot(x=main_primary_fuel,y=main_primary_fuel.index)
plt.title('Main primary fuel')
plt.xlabel('Count(Percentages)');
plt.ylabel('Different type of power plant deoends on the type of fuel uses')
```

# Type of powerplant and their capacity

```python
sns.scatterplot(x=df1.capacity_mw, y= df1.primary_fuel, s=150)
plt.title('Type of power plant and capacity');
```

```python
#now by using groupby() method we can split the dataframeinto name and their capacity of generating power
```

```python
states_capacity=df1.groupby('name')[['capacity_mw']].sum().sort_values('capacity_mw',ascending=False).head(20)
states_capacity
```

```python
sns.barplot(x=states_capacity.index,y= states_capacity.capacity_mw)
plt.xticks(rotation=90)
plt.title('States with capacity');
```
